```python
import pandas as pd

df = pd.read_csv(r"C:\Users\egorp\Desktop\survey_data_updated 5.csv")

multi_cols = [
    "LanguageHaveWorkedWith",
    "DatabaseHaveWorkedWith",
    "PlatformHaveWorkedWith",
    "WebframeHaveWorkedWith",
    "LanguageWantToWorkWith",
    "DatabaseWantToWorkWith",
    "PlatformWantToWorkWith",
    "WebframeWantToWorkWith"
]


for col in multi_cols:
    temp = df[[col]].copy()
    temp[col] = temp[col].fillna("").str.replace(" ", "")
    temp[col] = temp[col].str.split(";")
    temp = temp.explode(col)
    temp = temp[temp[col] != ""]

    temp.to_csv(fr"C:\Users\egorp\Desktop\cleaned_{col}.csv", index=False)

print("Saved to Desktop.")
```
